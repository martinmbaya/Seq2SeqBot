Gensim's LineSentence class provides a convenient iterator for working with other gensim components. It streams the documents/sentences from disk, so that you never have to hold the entire corpus in RAM at once. This allows you to scale your modeling pipeline up to potentially very large corpora.

```python
os.chdir("../Cleaned Data")
sentences = LineSentence("conversationData.txt")
model = Word2Vec(sentences, size=1, window=5, min_count=0, iter=100, workers=4)
# print('Saving the word embedding matrix')
# fname = "embeddingMatrix.npy"
# model.save(fname)
```

```python
# float(max(model.wv.vectors)+1)
# float(min(model.wv.vectors)-1)
list(model.wv.vectors)
mylist = []
mydic = {}
for i in list(model.wv.vectors):
    mylist.append(i[0])
mylist.sort()
count = 0
for i in mylist:
    mydic[float(i)] = count
    count += 1
print(mydic)
```
